### bigquery 연동하기 

```python
#bigquery 연동하기 
from google.oauth2 import service_account
from google.cloud import bigquery
```

```python
credentials = service_account.Credentials.from_service_account_file("./bigquery_key_json.json")
```

```python
client = bigquery.Client(
    credentials=credentials,
    project=credentials.project_id
)
```

### 지도시각화 서울 노인보호구역 

```python
# 서울 노인보호구역 21~24년자료
import pandas as pd
import folium

query = """
SELECT 
    facility, 
    latitude, 
    longitude

FROM `moonlit-vine-428804-p9.semi_project.old_safety_zone_1`

WHERE
    province = "서울특별시"

"""
```

```python
old_safety_zone_df = client.query(query).to_dataframe()
```

```python
# 노인 보호구역 21~24년 데이터 마커 추가
for idx, row in old_safety_zone_df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        popup=row['facility'],
        radius=5,
        color='blue',
        fill=True,
        fill_color='blue'
    ).add_to(map_positron)
```

### 보호소 주변 500M 내 사고 

```python
# 보호소 주변 500M 내 노인 보행 사고 자료 2012~2022
sql = """
SELECT 

    spot_nm AS spot_nm,
    la_crd AS latitude,
    lo_crd AS longitude

FROM `moonlit-vine-428804-p9.semi_project.by_region_accidents_within_500m` 

WHERE
  si = "서울특별시"
"""
```

```python
accidents_within_500m_df = client.query(sql).to_dataframe()
```

```python
# 보호소 주변 500M 내 노인 보행 사고 자료 2012~22
for idx, row in accidents_within_500m_df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        popup=row['spot_nm'],
        radius=5,
        color= 'red',
        fill=True,
        fill_color='red' # 두 번째 테이블의 마커 색상 지정
    
    ).add_to(map_positron)
```
